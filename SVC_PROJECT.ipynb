```python
def Total_Score(df):
  df['Total_Score'] = df['math score'] + df['reading score'] + df['writing score']
  if df['Total_Score'] > 250:
    df['Student_Level'] = '5'
  elif df['Total_Score'] > 200:
    df['Student_Level'] = '4'
  elif df['Total_Score'] > 150:
    df['Student_Level'] = '3'
  elif df['Total_Score'] > 100:
    df['Student_Level'] = '2'
  else:
    df['Student_Level'] = '1'
  return df
df['Total_Score'] = df.apply(Total_Score, axis=1)['Total_Score']
```

```python
df.head()
```

```python
df['Total_Score'].unique()
```

```python
def Level(Total_Score):
  if Total_Score > 250:
    return '5'
  elif Total_Score > 200:
    return '4'
  elif Total_Score > 150:
    return '3'
  elif Total_Score > 100:
    return '2'
  else:
    return '1'
df['Level'] = df['Total_Score'].apply(Level)
```

```python
df.head()
```

```python
df['Level'].unique()
```

```python
colors = {
    '1': 'green',
    '2': 'blue',
    '3': 'red',
    '4': 'orange',
    '5': 'purple'
}

plt.figure(figsize=(8, 5))

for cat in colors:
  data = df[df['Level'] == cat]
  plt.scatter(
      data.index,
      data['Total_Score'],
      color = colors[cat],
      label = cat,
      s=60
  )

plt.xlabel('Level')
plt.ylabel('Total_Score')
plt.title('Student Percentage by Level')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
x = df.drop(['Level'], axis=1)
y = df['Level']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,r2_score,mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x = pd.get_dummies(x, drop_first=True)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model = SVC()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
import matplotlib.pyplot as plt

def calculate_grade(total_score):
  if total_score > 250:
    return '5'
  elif total_score > 200:
    return '4'
  elif total_score > 150:
    return '3'
  elif total_score > 100:
    return '2'
  else:
    return '1'

plt.figure(figsize=(8,6))
colors = {'1': 'green','2': 'blue','3': 'red', '4': 'orange', '5': 'purple'}
grade_labels = {'1': 'Level A','2': 'Level B','3': 'Level C', '4': 'Level D', '5': 'Level E'}

for grade_numeric, color in colors.items():
    subset = df[df["Level"] == grade_numeric]
    plt.scatter(subset["math score"], subset["Total_Score"], color=color,label=grade_labels[grade_numeric],s=50)


input_overall_score = 70.0
student_level = calculate_grade(input_overall_score)

plt.scatter(
    input_overall_score,
    input_overall_score, # Assuming input_overall_score represents both math and total score for the hypothetical student
    color=colors[student_level],
    marker='X',
    s=300,
    edgecolor='black',
    linewidth=2,
    label=f'Input Student (Level {student_level})'
)

plt.title("Student Classification using SVM")
plt.xlabel("Math Score")
plt.ylabel("Overall Score")
plt.legend(title="Levels")
plt.grid(True)
plt.xlim(0,100)
plt.ylim(0,300) # Adjusted y-limit to accommodate Total_Score which can go up to 300

plt.show()
```
